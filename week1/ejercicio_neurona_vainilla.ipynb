## 1. Código base

Una neurona con función de activación *step* (umbral): devuelve 1 si \(z \geq 0\), sino 0.
Fórmula: \(z = x_1 w_1 + x_2 w_2 + b\)

```python
import numpy as np

def step(z):
    return 1 if z >= 0 else 0

def neurona(x1, x2, w1, w2, b):
    z = x1*w1 + x2*w2 + b
    y = step(z)
    return z, y

# Valores base
w1, w2, b = 1, 1, -1.5   # intenta parecerse a una compuerta AND
casos = [(0,0),(0,1),(1,0),(1,1)]

print("=== Código base: w1=1, w2=1, b=-1.5 ===")
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b)
    print(f"Entrada=({x1},{x2})  z={z:.1f}  salida={y}")
```

## 2. Pruebas cambiando solo el bias (b)

Mantenemos **w1=1, w2=1** y variamos **b** para ver cómo afecta la salida.

### Prueba 1: b = -0.5

Con **b = -0.5** el umbral es más "fácil" de superar: hace falta menos suma ponderada para que \(z \geq 0\). Por tanto, más combinaciones de entradas darán salida 1. Suele alejarse del comportamiento AND estricto (solo (1,1) → 1).

```python
w1, w2, b = 1, 1, -0.5
casos = [(0,0),(0,1),(1,0),(1,1)]

print("b = -0.5")
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b)
    print(f"Entrada=({x1},{x2})  z={z:.1f}  salida={y}")
```

### Prueba 2: b = -1.0

Con **b = -1.0** el umbral está entre -0.5 y -1.5. Para \(z \geq 0\) necesitamos \(x_1 + x_2 \geq 1\). Así (0,1) y (1,0) dan z=0 → salida 1, y (1,1) da z=1 → salida 1. Sigue sin ser AND puro (donde solo (1,1) debe dar 1), pero se acerca un poco más al control del umbral.

```python
w1, w2, b = 1, 1, -1.0
casos = [(0,0),(0,1),(1,0),(1,1)]

print("b = -1.0")
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b)
    print(f"Entrada=({x1},{x2})  z={z:.1f}  salida={y}")
```

### Prueba 3: b = -2.0

Con **b = -2.0** el umbral es más "exigente": hace falta que \(x_1 + x_2\) sea mayor para que \(z \geq 0\). Con w1=w2=1, solo cuando \(x_1+x_2 \geq 2\) (es decir, solo (1,1)) se tiene \(z \geq 0\). Por tanto la neurona se comporta como una compuerta **AND**.

```python
w1, w2, b = 1, 1, -2.0
casos = [(0,0),(0,1),(1,0),(1,1)]

print("b = -2.0")
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b)
    print(f"Entrada=({x1},{x2})  z={z:.1f}  salida={y}")
```

## 3. Preguntas y respuestas

### • ¿Con qué valor de b la neurona se comporta más parecido a AND?

Con **b = -2.0** (y w1=1, w2=1). En ese caso solo la entrada (1,1) da z ≥ 0 y por tanto salida 1; (0,0), (0,1) y (1,0) dan z < 0 y salida 0. Eso coincide exactamente con la tabla de verdad de la compuerta AND.

---

### • ¿Qué efecto tiene bajar o subir el bias en la salida? (1-2 líneas)

**Subir el bias** (acercarlo a 0 o positivo): la neurona "se activa más fácil"; más entradas producen z ≥ 0 y por tanto más salidas 1. **Bajar el bias** (más negativo): la neurona "exige más" para activarse; hace falta una suma ponderada mayor (p. ej. solo (1,1) con w1=w2=1) para que z ≥ 0, así que hay menos salidas 1. El bias actúa como un umbral que desplaza la frontera de decisión.
